```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('task3.csv', delimiter=',', skip_header=1)

fig, ax = plt.subplots(1, 2)
fig.suptitle('References vs Bitrate')
fig.set_size_inches(15, 5)

ax[0].plot(data[:, 1], data[:, 2]/1000.0, label='KristenAndSara', color='r')
ax[0].set_xlabel('References')
ax[0].set_ylabel('Bitrate (kbpbs)')

ax[1].plot(data[:, 1], data[:, 3]/1000.0, label='Stockholm', color='b')
ax[1].set_xlabel('References')
ax[1].set_ylabel('Bitrate (kbps)')

fig.legend()
fig.savefig('task3.png')
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('task6.csv', delimiter=',', skip_header=1)

fig, ax = plt.subplots(1, 2)
fig.suptitle('H.264 vs H.265')
fig.set_size_inches(15, 5)

ax[0].plot(data[:, 1]/1000.0, data[:, 2], label='H.264', color='b', linestyle='-.')
ax[0].plot(data[:, 3]/1000.0, data[:, 4], label='H.265', color='r', linestyle='-.')
ax[1].plot(data[:, 5]/1000.0, data[:, 6], label='H.264', color='b', linestyle='-.')
ax[1].plot(data[:, 7]/1000.0, data[:, 8], label='H.265', color='r', linestyle='-.')
ax[0].set_title('KristenAndSara_720p')
ax[1].set_title('Stockholm_720p')
ax[0].set_xlabel('Bitrate (kbps)')
ax[1].set_xlabel('Bitrate (kbps)')
ax[0].set_ylabel('PSNR (dB)')
ax[1].set_ylabel('PSNR (dB)')
ax[0].set_ybound(20, 50)
ax[1].set_ybound(20, 50)

ax[0].legend()
ax[1].legend()
fig.savefig('task6.png')
plt.show()
```

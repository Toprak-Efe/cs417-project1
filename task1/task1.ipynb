```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def bitrate_to_kbps(bitrate):
    return bitrate / 1000
```

```python
# Read test1.csv
data = np.genfromtxt('task1.csv', delimiter=',', skip_header=1)
data[:, 0] = data[:, 0]*10

# Plot the data
fig, ax = plt.subplots(1, 3)
fig.suptitle('Effect of QP on Bitrate')
fig.set_size_inches(15, 5)

ax[0].set_title('QP vs Bitrate')
ax[0].set_xlabel('QP')
ax[0].set_ylabel('Bitrate (kbps)')
ax[0].plot(data[:, 0], data[:, 1]/1000.0, color='r', label='KristenAndSara_720p')
ax[0].plot(data[:, 0], data[:, 4]/1000.0, color='g', label='Stockholm_720p')
ax[0].set_yscale('log')
ax[0].set_ylim(0, 10**5)
ax[0].legend(loc='upper right')

ax[1].set_title('QP vs PSNR')
ax[1].set_xlabel('QP')
ax[1].set_ylabel('PSNR')
ax[1].plot(data[:, 0], data[:, 2], color='r', label='KristenAndSara_720p')
ax[1].plot(data[:, 0], data[:, 5], color='g', label='Stockholm_720p')
ax[1].set_ylim(0, 60)
ax[1].legend(loc="upper right")

ax[2].set_title('QP vs SSIM')
ax[2].set_xlabel=('QP')
ax[2].set_ylabel=('SSIM')
ax[2].plot(data[:, 0], data[:, 3], color='r', label='KristenAndSara_720p')
ax[2].plot(data[:, 0], data[:, 6], color='g', label='Stockholm_720p')
ax[2].set_ylim(0, 1)
ax[2].legend(loc="upper right")

# Save figure to file.
fig.savefig('task1.png')
plt.show()
```

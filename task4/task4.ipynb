```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('task4.csv', delimiter=',', skip_header=1)

fig, ax = plt.subplots(2, 2)
fig.suptitle('Motion Estimation and Bitrate')
fig.set_size_inches(10, 10)
fig.tight_layout(pad=3.0)

indices = [3, 7, 15]

len = data.shape[0]
dia = data[:len//2, 1:]
esa = data[len//2:, 1:]

#retain only the rows with the indices
range = [i+1 for i in indices]
dia = dia[indices]
esa = esa[indices]

ax[0, 0].plot(range, dia[:, 1]/1000.0, label='DIA', color='r')
ax[0, 0].set_title('KristenAndSara')
ax[0, 0].set_ylabel('Bitrate (kbps)')
ax[0, 0].set_xlabel('Motion Estimation Range')
ax[0, 0].legend()
ax[1, 0].plot(range, esa[:, 1]/1000.0, label='ESA', color='r')
ax[1, 0].set_ylabel('Bitrate (kbps)')
ax[1, 0].set_xlabel('Motion Estimation Range')
ax[1, 0].legend()

ax[0, 1].plot(range, dia[:, 2]/1000.0, label='DIA', color='g')
ax[0, 1].set_title('Stockholm')
ax[0, 1].set_ylabel('Bitrate (kbps)')
ax[0, 1].set_xlabel('Motion Estimation Range')
ax[0, 1].legend()
ax[1, 1].plot(range, esa[:, 2]/1000.0, label='ESA', color='g')
ax[1, 1].set_ylabel('Bitrate (kbps)')
ax[1, 1].set_xlabel('Motion Estimation Range')
ax[1, 1].legend()

fig.legend(['KristenAndSara', 'Stockholm'])
fig.show()
fig.savefig('task4.png')
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('task5.csv', delimiter=',', skip_header=1)

fig, ax = plt.subplots(1, 2)
fig.suptitle('Motion Estimation and Bitrate')
fig.set_size_inches(15, 5)

ax[0].set_title('KristenAndSara')
ax[0].plot(data[:, 0], data[:, 1]/1000.0, label='DIA', color='r')
ax[0].set_ylabel('Bitrate (kbps)')
ax[0].set_xlabel('BFrames')

ax[1].set_title('Stockholm')
ax[1].plot(data[:, 0], data[:, 1]/1000.0, label='DIA', color='g')
ax[1].set_ylabel('Bitrate (kbps)')
ax[1].set_xlabel('BFrames')

fig.show()
fig.savefig('task5.png')
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('task2.csv', delimiter=',', skip_header=1)

fig, ax = plt.subplots()
fig.suptitle('GOP Length vs Bitrate')

ax.plot(data[:, 1], data[:, 2]/1000.0, label='KristenAndSara')
ax.plot(data[:, 1], data[:, 3]/1000.0, label='Stockholm')
ax.set_xlabel('GOP Length')
ax.set_ylabel('Bitrate (kbps)')
ax.set_yscale('log')
ax.set_ybound(0, 100000)
fig.legend()

fig.savefig('task2.png')
```
